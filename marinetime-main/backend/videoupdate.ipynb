```python
import cv2
import os

input_path = r"D:\marinetime\backend\systemvideo\video12.mp4"
output_path = r"systemvideo\video12_10min_10fps.mp4"

target_fps = 10
target_duration_minutes = 10
target_total_frames = target_fps * 60 * target_duration_minutes  # 6000


def main():
    if not os.path.exists(input_path):
        print("Input file not found:", input_path)
        return

    cap = cv2.VideoCapture(input_path)

    if not cap.isOpened():
        print("Failed to open input video")
        return

    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    source_frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if width <= 0 or height <= 0:
        print("Invalid video size")
        cap.release()
        return

    if source_frame_count <= 0:
        print("Invalid source frame count")
        cap.release()
        return

    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out = cv2.VideoWriter(output_path, fourcc, target_fps, (width, height))

    if not out.isOpened():
        print("Failed to create output video")
        cap.release()
        return

    print("Input:", input_path)
    print("Output:", output_path)
    print("Resolution:", width, "x", height)
    print("Source frames:", source_frame_count)
    print("Target frames:", target_total_frames)
    print("Target FPS:", target_fps)

    written = 0

    while written < target_total_frames:
        frame_index = written % source_frame_count

        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_index)
        ret, frame = cap.read()

        if not ret or frame is None:
            print(f"Failed reading frame at index {frame_index}")
            break

        out.write(frame)
        written += 1

        if written % 100 == 0:
            print(f"Written {written}/{target_total_frames} frames")

    cap.release()
    out.release()

    print("Done")
    print("Frames written:", written)
    print("Saved to:", output_path)


if __name__ == "__main__":
    main()
```

```python
import cv2
import os

video_files = [
    r"systemvideo\video12.mp4",
    r"systemvideo\video13.mp4",
    r"systemvideo\video14.mp4",
    r"systemvideo\video15.mp4",
    r"systemvideo\video16.mp4",
]

output_path = r"systemvideo\combined.mp4"
target_fps = 20


def main():
    # Check files
    for f in video_files:
        if not os.path.exists(f):
            print("Missing file:", f)
            return

    # Open first video to get properties
    cap0 = cv2.VideoCapture(video_files[0])

    if not cap0.isOpened():
        print("Failed to open first video")
        return

    width = int(cap0.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap0.get(cv2.CAP_PROP_FRAME_HEIGHT))
    cap0.release()

    # Create writer
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out = cv2.VideoWriter(output_path, fourcc, target_fps, (width, height))

    if not out.isOpened():
        print("Failed to create output video")
        return

    total_written = 0

    for video in video_files:
        print("Processing:", video)
        cap = cv2.VideoCapture(video)

        if not cap.isOpened():
            print("Skipping:", video)
            continue

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Resize if needed (important if sizes differ)
            frame = cv2.resize(frame, (width, height))

            out.write(frame)
            total_written += 1

        cap.release()

    out.release()

    print("Done!")
    print("Total frames written:", total_written)
    print("Saved to:", output_path)


if __name__ == "__main__":
    main()
```
